# Customer Personality Analysis - Preprocessing 

```python
import pandas as pd

df = pd.read_csv('marketing_campaign.csv', sep='\t')
df.drop('ID', axis=1, inplace=True)

df.head()
```

## Itinerario


###  Personal
1. Eliminar columnas: ID, Dt_Customer, Complain, Z_CostContact, Z_Revenue 
2. Elimar los pocos outliers de INCOME
3. Modificar los valores mas desbalanceados d Marital_Status y llevarlos a una categoria parecida
4. Modificar columnas: Teenhome, Kidhnome (modificacion a valores binarios para evitar el desbalanceo)


```python
# 1.
df.drop('Dt_Customer', axis=1, inplace=True)
df.drop('Complain', axis=1, inplace=True)
df.drop('Z_CostContact', axis=1, inplace=True)
df.drop('Z_Revenue', axis=1, inplace=True)

df.head(2)
```

```python
# Datos con las categorias minoritarias de 'Marital_Status'
df[['Income', 'Marital_Status']][df['Marital_Status'].isin(['YOLO', 'Alone', 'Absurd'])]
```
